```python
# Importing libraries
from faker import Faker
import pandas as pd
import numpy as np
import random

# Initialize Faker to generate fake data
fake = Faker()

# Generate fake data for the DataFrame
data = {
    'Year': [str(year) for year in range(2000, 2024)],
    'Total_Cybercrimes': [random.randint(10000, 50000) for _ in range(24)],
    'Financial_Fraud': [random.randint(2000, 10000) for _ in range(24)],
    'Identity_Theft': [random.randint(1000, 8000) for _ in range(24)],
    'Cyberbullying': [random.randint(500, 3000) for _ in range(24)],
    'Hacking': [random.randint(200, 5000) for _ in range(24)],
    'Data_Breaches': [random.randint(500, 4000) for _ in range(24)],
    'Economic_Cybercrimes': [random.randint(1000, 10000) for _ in range(24)],
    'Online_Scams': [random.randint(500, 2000) for _ in range(24)],
    'Phishing_Attacks': [random.randint(200, 1000) for _ in range(24)],
    'Ransomware_Attacks': [random.randint(100, 500) for _ in range(24)],
    'Malware_Infections': [random.randint(500, 3000) for _ in range(24)],
    'Social_Media_Fraud': [random.randint(200, 1000) for _ in range(24)],
    'Dark_Web_Activity': [random.randint(100, 500) for _ in range(24)],
    'Cybersecurity_Breaches': [random.randint(500, 2000) for _ in range(24)],
    'Cases_Charge_Sheeted': [random.randint(200, 1000) for _ in range(24)],
    'Persons_Arrested': [random.randint(500, 3000) for _ in range(24)]
}

# Create DataFrame
df = pd.DataFrame(data)

# Reorder the columns
df = df[['Year', 'Total_Cybercrimes', 'Financial_Fraud', 'Identity_Theft', 
         'Cyberbullying', 'Hacking', 'Data_Breaches', 'Economic_Cybercrimes', 
         'Online_Scams', 'Phishing_Attacks', 'Ransomware_Attacks', 'Malware_Infections', 
         'Social_Media_Fraud', 'Dark_Web_Activity', 'Cybersecurity_Breaches',
         'Cases_Charge_Sheeted', 'Persons_Arrested']]

# Add NaN values for the years 2000 to 2023 
nan_years = random.sample(range(2000, 2024), 5)  # Randomly choose 5 years
for year in nan_years:
    df.loc[df['Year'] == str(year), df.columns[2:]] = np.nan

# Add null values for the years 2000 to 2023 (excluding 'State' column)
null_years = random.sample(range(2000, 2024), 5)  # Randomly choose 5 years
for year in null_years:
    df.loc[df['Year'] == str(year), df.columns[2:]] = None

# Add duplicated values for the years 2000 to 2023 (excluding 'State' column)
dup_years = random.sample(range(2000, 2024), 5)  # Randomly choose 5 years
for year in dup_years:
    df = pd.concat([df, df[df['Year'] == str(year)]], ignore_index=True)

# Display the DataFrame
print(df)

```

```python
# To perform a trend analysis of various cybercrime categories over multiple years, excluding the total cybercrimes.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame containing the data

# Select only the desired columns for analysis
columns_for_analysis = df[['Year','Financial_Fraud', 'Identity_Theft', 'Cyberbullying','Data_Breaches',
                           'Economic_Cybercrimes', 'Online_Scams', 'Phishing_Attacks', 
                           'Malware_Infections', 'Social_Media_Fraud', 'Dark_Web_Activity', 'Cybersecurity_Breaches']]

# Group by year and calculate mean for each column
yearly_mean = columns_for_analysis.groupby('Year').mean()

# Plot bar chart for each column
plt.figure(figsize=(12, 8))
for column in yearly_mean.columns:
    plt.bar(yearly_mean.index, yearly_mean[column], label=column)
plt.xlabel('Year')
plt.ylabel('Mean Value')
plt.title('Trend Analysis of Cybercrime Data')
plt.legend()
plt.show()

```

```python
#State-wise Analysis
# Calculate total cybercrimes for each state
statewise_total = df.groupby('States')['Total_Cybercrimes'].sum().reset_index()

# Sort states by total cybercrimes in descending order
statewise_total = statewise_total.sort_values(by='Total_Cybercrimes', ascending=False)

# Plot bar chart for total cybercrimes by state
plt.figure(figsize=(10, 6))
plt.plot(statewise_total['States'], statewise_total['Total_Cybercrimes'],marker='o')
plt.xlabel('State')
plt.ylabel('Total Cybercrimes')
plt.title('Total Cybercrimes by State')
plt.xticks(rotation=90)
plt.show()
```

```python
#Correlation Matrix
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Assuming 'df' is your DataFrame containing the data

# Select only the desired columns for correlation analysis
columns_for_correlation = df[['Financial_Fraud', 'Identity_Theft', 'Cyberbullying', 'Data_Breaches',
                              'Economic_Cybercrimes', 'Online_Scams', 'Phishing_Attacks',
                              'Malware_Infections', 'Social_Media_Fraud', 'Dark_Web_Activity',
                              'Cybersecurity_Breaches']]

# Calculate correlation matrix for selected columns
correlation_matrix = columns_for_correlation.corr()

# Plot pie chart for each column's correlation with the first column
plt.figure(figsize=(12, 8))
first_column_correlation = correlation_matrix.iloc[:, 0]  # Select correlations with the first column
labels = first_column_correlation.index
plt.pie(np.abs(first_column_correlation), labels=labels, autopct='%1.1f%%', startangle=140)       #The absolute value of a number is its distance from 0 on the number line, without considering its direction (positive or negative). This function can handle both real and complex numbers.
plt.title('Correlation of Selected Columns with the First Column')
plt.show()
```

```python
#Arrests and Charge Sheet Analysis 
# Plot scatter plots for arrests and charge sheets vs. total cybercrimes
plt.figure(figsize=(10, 6))
plt.scatter(df['Total_Cybercrimes'], df['Cases_Charge_Sheeted'], label='Cases Charge Sheeted')
plt.scatter(df['Total_Cybercrimes'], df['Persons_Arrested'], label='Persons Arrested')
plt.xlabel('Total Cybercrimes')
plt.ylabel('Number of Cases/Persons')
plt.title('Arrests and Charge Sheets vs. Total Cybercrimes')
plt.legend()
plt.grid(True)
plt.show()
```

```python
#Time-Series Analysis
# Convert the index to a DatetimeIndex
df.index = pd.to_datetime(df.index)

# Plot box plots for each column
plt.figure(figsize=(12, 6))
df.boxplot()
plt.xlabel('Variable')
plt.ylabel('Value')
plt.title('Box Plot of Time Series Data')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.grid(True)  # Add grid lines for better readability
plt.show()
```
